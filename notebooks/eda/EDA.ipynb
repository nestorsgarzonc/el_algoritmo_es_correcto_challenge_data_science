```
def filter_by_quantile(df: pd.DataFrame, name_col: str, min = 0, max = 0.9999):
  min_val = df[name_col].quantile(min)
  max_val = df[name_col].quantile(max)
  return df[(df[name_col] >= min_val) & (df[name_col] <= max_val)]
```
