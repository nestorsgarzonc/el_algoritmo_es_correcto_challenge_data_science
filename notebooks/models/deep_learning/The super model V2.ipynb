## Add onehot encoding to category variables

```python
def AddOnehotEncoding(df):
  
  try:
    one_hot_magic={
      '0-1 anos': ['Menos de 1 aÃ±o'],
      '1-10 anos': ['ntre 0 y 5 aÃ±os', 'Entre 0 y 5 aÃ±os', '1 a 8 aÃ±os', 'Entre 5 y 10 aÃ±os'],
      '10-30 anos': ['9 a 15 aÃ±os', 'Entre 10 y 20 aÃ±os', '16 a 30 aÃ±os', 'MÃ¡s de 20 aÃ±os'],
      '30 anos': ['MÃ¡s de 30 aÃ±os'],
    }
    d = []
    for k, v in one_hot_magic.items():
      a = pd.Series([])
      for y in v:
        if len(df.value_counts()) <= 0:
          a=df.tiempodeconstruido == y
        else:
          a= a | df.tiempodeconstruido == y
      df[k] = a
      df[k]=df[k].replace(True, 1).replace(False, 0).astype(int)
  except Exception as e:
    print(e) 
  tiponegocio = pd.get_dummies(df.tiponegocio, prefix='tiponegocio')
  columns = tiponegocio.columns

  df['tiponegocio_Arriendo'] = tiponegocio['tiponegocio_Arriendo'] if 'tiponegocio_Arriendo' in columns else 0
  df['tiponegocio_Venta'] = tiponegocio['tiponegocio_Venta'] if 'tiponegocio_Venta' in columns else 0
  df['tiponegocio_Venta Y Arriendo'] = tiponegocio['tiponegocio_Venta Y Arriendo']+tiponegocio['tiponegocio_Venta y arriendo']	if 'tiponegocio_Venta Y Arriendo' in columns and 'tiponegocio_Venta y arriendo' in columns else 0
  df.pop('tiponegocio')

  tipoinmueble  = pd.get_dummies(df.tipoinmueble , prefix='tipoinmueble')

  df.pop('tipoinmueble')
  columns = tipoinmueble.columns
  df['tipoinmueble_Casa'] = tipoinmueble['tipoinmueble_Casa'] if 'tipoinmueble_Casa' in columns else 0
  df['tipoinmueble_Apartamento'] = tipoinmueble['tipoinmueble_Apartamento'] if 'tipoinmueble_Apartamento' in columns else 0   
  return df
```

```python
df = AddOnehotEncoding(df)
```
