## Add onehot encoding to category variables

```python
def AddOnehotEncoding(df):
  
  try:
    one_hot_magic={
      '0-1 anos': ['Menos de 1 año'],
      '1-10 anos': ['ntre 0 y 5 años', 'Entre 0 y 5 años', '1 a 8 años', 'Entre 5 y 10 años'],
      '10-30 anos': ['9 a 15 años', 'Entre 10 y 20 años', '16 a 30 años', 'Más de 20 años'],
      '30 anos': ['Más de 30 años'],
    }
    d = []
    for k, v in one_hot_magic.items():
      a = pd.Series([])
      for y in v:
        if len(df.value_counts()) <= 0:
          a=df.tiempodeconstruido == y
        else:
          a= a | df.tiempodeconstruido == y
      df[k] = a
      df[k]=df[k].replace(True, 1).replace(False, 0).astype(int)
  except Exception as e:
    print(e) 
  tiponegocio = pd.get_dummies(df.tiponegocio, prefix='tiponegocio')
  columns = tiponegocio.columns

  df['tiponegocio_Arriendo'] = tiponegocio['tiponegocio_Arriendo'] if 'tiponegocio_Arriendo' in columns else 0
  df['tiponegocio_Venta'] = tiponegocio['tiponegocio_Venta'] if 'tiponegocio_Venta' in columns else 0
  df['tiponegocio_Venta Y Arriendo'] = tiponegocio['tiponegocio_Venta Y Arriendo']+tiponegocio['tiponegocio_Venta y arriendo']	if 'tiponegocio_Venta Y Arriendo' in columns and 'tiponegocio_Venta y arriendo' in columns else 0
  df.pop('tiponegocio')

  tipoinmueble  = pd.get_dummies(df.tipoinmueble , prefix='tipoinmueble')

  df.pop('tipoinmueble')
  columns = tipoinmueble.columns
  df['tipoinmueble_Casa'] = tipoinmueble['tipoinmueble_Casa'] if 'tipoinmueble_Casa' in columns else 0
  df['tipoinmueble_Apartamento'] = tipoinmueble['tipoinmueble_Apartamento'] if 'tipoinmueble_Apartamento' in columns else 0   
  return df
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
def calculateCV(myNewDf, numLabels): 
  cvs = []
  for f in range(numLabels):
    prueba = myNewDf[myNewDf['Label k-means'] == f].describe()
    cv = prueba.loc['std', 'val_mcuadrado'] / prueba.loc['mean', 'val_mcuadrado'] * 100
    cvs.append(cv)
  print("for", numLabels)
  print("we have:", cvs)
  print("MAX cv:", max(cvs))
  return cvs
```

```python
def km2df(kmeans, df):
  dataLabels = pd.Series(kmeans.labels_)
  dataLabels.index = df.index
  myNewDf = pd.concat((df, dataLabels), axis=1)
  myNewDf = myNewDf.rename(columns = {0:'Label k-means'})
  return myNewDf
```

```python
def kmeansLearning(kmax, x_data, kmeans_models):
  # for n_clusters in range()
  for n_clusters in range(5, kmax + 1):
    kmeans = KMeans(n_clusters=n_clusters, n_init = 20, n_jobs=-1, verbose = 1)
    kmeans.fit(x_data)
    kmean_df = km2df(kmeans)    
    calculateCV(kmean_df, n_clusters)
    kmeans_models.append(kmeans)
    print("se añadió el modelo")
  return kmeans_models
```

```python
def calculate_WSS(kmeans_models, points):
  x_data = np.array(points)
  sse = []
  count = 5
  for kmeans in kmeans_models:
    print("listo el ", count)
    count += 1
    centroids = kmeans.cluster_centers_
    pred_clusters = kmeans.predict(x_data)
    curr_sse = 0
    
    # calculate square of Euclidean distance of each point from its cluster center and add to current WSS
    for i in range(x_data.shape[0]):
      curr_center = np.array(centroids[pred_clusters[i]])
      curr_sse += np.sum((x_data[i] - curr_center)**2)  
    sse.append(curr_sse)
  return sse
```

```python
def calculate_silhouette(kmeans_models, points):
  scs = []
  for kmeans in kmeans_models:
    sc = silhouette_score(points, kmeans.labels_, metric = 'euclidean', n_jobs = -1)
    scs.append(sc)
  return scs
```

```python
sse = calculate_WSS(kmeans_models, x_data)
```

```python
def accesCVPredictions(kmeans, x_data, y_data):
  myDf = km2df(kmeans, x_data)
  cv = calculateCV(myDf, kmeans.n_clusters)
  setA = set(kmeans.predict(y_data))
  resp = [cv[x] for x in setA]
  return resp, np.mean(resp)
```
